## 세트권 복제

```python
# 파일이 3곳에 나뉘어져있으니, 복사하는 함수로 진행
def find_img(ge_file, code, set_code):
    if code != set_code:
        dst = f'/content/drive/MyDrive/yongga/image/{code}.jpg'
        shutil.copyfile(ge_file, dst)
        print('개별권 복제완료')
    else:
        print('세트권')
```

```python
mu,hand = [],[]
for i in range(len(df)):
    code,set_code =df['isbn'][i], df['set_isbn'][i]
    img_file = f'/content/drive/MyDrive/yongga/image/{code}.jpg'
    set_file = f'/content/drive/MyDrive/yongga/image/{set_code}.jpg'
    if os.path.exists(img_file): # 이미지 파일이 존재할때
        if set_code != 0: # 세트의 세트권 이미지 존재
            pass
        else:
            print(code,'존재')
    elif os.path.exists(set_file):
        if set_code != 0: # 세트 이미지 데이터 존재
            find_img(set_file, code, set_code)
            # hand.append(set_code)
    else:
        hand.append(code)
```
