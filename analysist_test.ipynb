```python
sepal_l = iris['sepal_length']
sepal_l_mean = np.mean(sepal_l)
sepal_w = iris['sepal_width']
sepal_w_mean = np.mean(sepal_w)
petal_l = iris['petal_length']
petal_l_mean = np.mean(petal_l)
petal_w = iris['petal_width']
petal_w_mean = np.mean(petal_w)
```

```python
# 단일표본 T-검정
from scipy.stats import ttest_1samp

tstat, pval = ttest_1samp(sepal_l,sepal_l_mean )
print(tstat, pval)
```

```python
# 독립표본 T-검정
from scipy.stats import ttest_ind

tstat, pval = ttest_ind(sepal_l, sepal_w)
print(tstat, pval)
```

```python
# 분산 분석
from scipy.stats import f_oneway

fstat, pval = f_oneway(sepal_l, sepal_w, petal_l)
print(fstat, pval)
```

```python
# 사후 검정
## Tukey's Range Test 방법
from scipy.stats import f_oneway
import numpy as np
from statsmodels.stats.multicomp import pairwise_tukeyhsd

fstat,pval = f_oneway(sepal_l, sepal_w, petal_l)

# np.concatenate로 데이터 합침
iris_datas = np.concatenate([sepal_l, sepal_w, petal_l])
labels = ['sepal_l']*len(sepal_l) + ['sepal_w']*len(sepal_w) + ['petal_l']*len(petal_l)

# 사후검정
tukey_results = pairwise_tukeyhsd(iris_datas, labels, 0.05)
print(tukey_results)  ## reject 귀무가설을 기각하라(비교그룹간 평균이 다르다)
```

- T-검정과 ANOVA의 전제 조건<br>
→ 분포의 평균을 비교하는 가설 검정임. 범주가 2개로 구성된 경우는 이항 검정을 사용해야함
> 정규성: 표본이 정규 분포를 그려야 한다.<br>
> 등분산성: 2개 이상의 집단을 비교하려면, 각 집단의 표준편차가 서로 같아야 한다.<br>
> 표본은 독립적이어야 하며, 서로 영향을 미치지 않아야 한다.
