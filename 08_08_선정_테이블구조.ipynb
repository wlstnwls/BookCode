```python
plusInfo = raw.loc[:, ('txtb_id', 'book_info','ref_info','book_review','ref_review')]
plusInfo
```

```python
def ref_plus(book, column, keyword):
    x = book[column]
    for i in range(len(x)):
        if x[i] == f"{keyword} 교보문고":
            x[i] = '교보문고'
        elif x[i] == f"{keyword} 알라딘에":
            x[i] = '알라딘'
        elif x[i] == f"{keyword} 예스24":
            x[i] = '예스24'
        else:
            x[i] = np.NaN
    return x
```

```python
column = 'ref_review'
keyword = '서평'

x = ref_plus(plusInfo,column, keyword)

plusInfo['ref_review'] = x
```

```python
column = 'ref_info'
keyword = '소개'

x = ref_plus(plusInfo,column, keyword)
x
plusInfo['ref_info'] = x
```

```python
plusInfo.columns = ['txtb_id','txtb_itr', 'txtb_itr_src', 'txtb_brevw', 'txtb_brevw_src']
plusInfo.head()
```

```python
def store_aut(aut,code):
    for a in aut:
        a = re.sub(r'[^a-zA-Z가-힣]','',a)
        aut_ls.append(a)
        code_ls.append(code)  
```

```python
code_ls, aut_ls = [],[]
for i in range(len(kjAuthor['isbn'])):
    code = kjAuthor['isbn'][i]
    aut = kjAuthor['author'][i]
    if type(aut) != float:
        if ',' in aut:
            aut = aut.split(',')
            store_aut(aut,code)
        elif '(' in aut:
            aut = aut.split(',')
            store_aut(aut,code)
        else:
            aut_ls.append(aut)
            code_ls.append(code)   
```

# kwd

```python
bookKeyword = raw.loc[:, ('txtb_id','cate','cate_kb','keyword')]
bookKeyword['cate_kb'] = bookKeyword['cate_kb'].replace(np.NaN, '')
bookKeyword['keyword'] = bookKeyword['keyword'].replace(np.NaN, '')
bookKeyword.head()
```

```python
isbn_ls, keyword_ls, ref_ls = [],[],[]

def keyword_split_kb(test,code, i):
    kbkey = re.findall(r"'(.*?)'", test[i])
    for kb in kbkey:
        isbn_ls.append(code[i])
        keyword_ls.append(kb)
        ref_ls.append('교보문고')
        
def keyword_split_isbn(task,code, i): 
    isbnkey = ''
    if ',' in task[i]:
        isbnkey = task[i].split(',')
    elif';' in task[i]:
        isbnkey = task[i].split(';')
    elif '|' in task[i]:
        isbnkey = task[i].split('|')
        
    isbnkey = [i for i in isbnkey if i != '']
    for ib in isbnkey:
        isbn_ls.append(code[i])
        keyword_ls.append(ib)
        ref_ls.append('국립중앙도서관')
```

```python
test = bookKeyword.loc[:,'cate_kb']
task = bookKeyword.loc[:,'keyword']
code = bookKeyword.loc[:,'txtb_id']
task.index = range(len(task))
code.index = range(len(code))
```

```python
for i in range(len(code)):
    if test[i] != '' and task[i] != '':
        keyword_split_kb(test,code, i)
        keyword_split_isbn(task,code, i)
#         isbn_ls.append(code[i])
    elif test[i] != '' and task[i] == '':
        keyword_split_kb(test,code, i)
#         isbn_ls.append(code[i])
    elif test[i] == '' and task[i] != '':
        keyword_split_isbn(task,code, i)
#         isbn_ls.append(code[i])
    else:
        isbn_ls.append(code[i])
        keyword_ls.append(np.NaN)
        ref_ls.append('nodata')
```

```python
dff = pd.DataFrame({'txtb_id':isbn_ls, 'kwd_nm':keyword_ls, '출처':ref_ls})
dff
```

```python
keyword_df = dff[['txtb_id','kwd_nm']]
keyword_df.info()
```

```python
keyword_df.dropna(inplace=True)
```

```python
keyword_df.info()
```

```python
keyword_df.to_excel('C:\practice_data\선정_키워드_08_09.xlsx',encoding = 'utf-8-sig', index=False)
```
